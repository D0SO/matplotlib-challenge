```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset using "Mouse ID" ad the 
data_set = pd.merge(mouse_metadata, study_results, on="Mouse ID")
# Display the data table for preview
data_set.head()
```

```python
# Checking the number of mice. Find number of unique id's in the Mouse ID column
data_set.nunique()["Mouse ID"]
```

```python
# Identify the ID of the mice with a duplicate entry on both Mouse ID and Timepoint. 
duplicate_mouse_id = data_set.loc[data_set.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicate_mouse_id
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_data = data_set.loc[data_set["Mouse ID"] == "g989"]
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# Exclude the duplicate ID from our clean data set
clean_data = data_set[data_set['Mouse ID'] != 'g989']

clean_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
# Group ourdata set by each unique mouse ID
unique_mice = clean_data.groupby(['Mouse ID']).nunique()
# Calculate the number of unique ID's
len(unique_mice)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

# Group our data byeach drug prescribed and perform our calculations on the "Tumor Volume (mm3)" column
tumor_mean = clean_data.groupby('Drug Regimen').mean()['Tumor Volume (mm3)'] 
tumor_median = clean_data.groupby('Drug Regimen').median()['Tumor Volume (mm3)'] 
tumor_variance = clean_data.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
tumor_std_dev = clean_data.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
tumor_sem = clean_data.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

# Create the labels for each column and insert our calculations
summary_table = pd.DataFrame({'Mean Tumor Volume': tumor_mean,
                             'Median Tumor Volume': tumor_median,
                             'Tumor Volume Variance': tumor_variance,
                             'Tumor Volume Std. Dev.': tumor_std_dev,
                             'Tumor Volme STd. Err.': tumor_sem})

summary_table
```

```python
# Alternatively we can acheive the same dataframe using a groupby function followed by an aggregation 
# function using the key terms "mean", "median", "var", "std", "sem"  for our columns 

new_summary_table = clean_data.groupby('Drug Regimen').agg({'Tumor Volume (mm3)': ["mean","median","var","std","sem"]})
new_summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_tumor = clean_data.groupby(['Mouse ID'])['Timepoint'].max()
# Reset the index values to the order of the new data frame
last_tumor = last_tumor.reset_index()
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merged_data = last_tumor.merge(clean_data, on = ['Mouse ID','Timepoint'], how = 'left')
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    last_tumor_volume = merged_data.loc[merged_data['Drug Regimen'] == drug, 'Tumor Volume (mm3)']
    
    # add subset 
    tumor_volume.append(last_tumor_volume)
    
    # Determine outliers using upper and lower bounds
    # Quartile measurement
    quartiles = last_tumor_volume.quantile([0.25,0.5,0.75])
    # lower quartile
    lower_q = quartiles[0.25]
    # Upper quartile
    wupper_q = quartiles[0.75]
    # Interquartile range (IQR)
    iqr = upper_q - lower_q
    # Create lower bound to identify outliers
    lower_bound = lower_q - (1.5*iqr)
    # Create higher bound to identify outliers
    upper_bound = upper_q + (1.5*iqr)
    # Identify tumor volume values that lay below its lower bound or above its upper bound
    outliers = last_tumor_volume.loc[(last_tumor_volume < lower_bound) | (last_tumor_volume > upper_bound)]  
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
# Create our boxplot using the values stored in tumor_volume variables using our 4 drug regimens as labels
plt.boxplot(tumor_volume, labels = treatments)
plt.ylabel('Final Tumor Volume (mm3)')
# Display graph with no additional information
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
# Select a mouse treated with Capomulin - s185 - exract its data frame
mouse_df = clean_data.loc[clean_data ['Mouse ID'] == 's185',:]
# Create a variable to store Timepoint values
time_point = mouse_df['Timepoint']
# Create a variable to store tumor volume values
tumor_volume = mouse_df['Tumor Volume (mm3)']
# Create our line chart
plt.plot(time_point, tumor_volume, linewidth = 2)
plt.title('Capolim tratment of mouse s185')
plt.xlabel('Timepoint (Days)')
plt.ylabel('Tumor Volume (mm3)')
# Display graph with no additional information
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
# Create a table that only contains Capomulin treated mice
capomulin = clean_data.loc[clean_data['Drug Regimen'] == 'Capomulin']
# The next table  groups our information per mouse ID and calculates the average for the remaining
# variables
capomulin_avg = capomulin.groupby(['Mouse ID']).mean()
# Create a scatter plot with Average Tumor Volume on y-axis and average weight on the x-axis
plt.scatter(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])
# Label our plot
plt.xlabel('Weight (g)')
plt.ylabel(' Average Tumor Volume (mm3)')
plt.title('Average Tumor Volume vs Mouse Weight')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
# calculate correlation coeficient
correl = round(st.pearsonr(capomulin_avg['Weight (g)'],capomulin_avg['Tumor Volume (mm3)'])[0],2)

reg_model = st.linregress(capomulin_avg['Weight (g)'],capomulin_avg['Tumor Volume (mm3)'])

# through our linear regression model identify the the slope and intercept so we can store
# the y function into a variable and plot it in the scatter plot 
slope = reg_model[0]
intercept = reg_model[1]
# Define our y function before we plot it
y_value = capomulin_avg['Weight (g)']*slope + intercept
# display the same scatter plot as before
plt.scatter(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])
# plot our linear regression line in the scatter plot
plt.plot(capomulin_avg['Weight (g)'],y_value, color = "black")
plt.xlabel('Weight (g)')
plt.ylabel(' Average Tumor Volume (mm3)')
plt.title('Average Tumor Volume vs Mouse Weight')
plt.show()
# Print the data of our directly calculated correlation and linear regression results
print(f"Correlation:{correl}")
print(f"Linear Regression Results")
print(f"R-Value: {reg_model[2]}")
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"P-Value: {reg_model[3]}")
print(f"stderr: {reg_model[4]}")
```
